```python
%aimport approxmh.vae
%aimport approxmh.distributions
%aimport approxmh.distribution_metrics
%aimport approxmh.y_utils
%aimport approxmh.samplers

import torch
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
from time import time
from tqdm import tqdm
from collections import defaultdict

from approxmh.y_utils import *
from approxmh.vae import VAE, get_filename
from approxmh.samplers import *
from approxmh.distributions import Banana, Funnel
from approxmh.distribution_metrics import *
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('device:', device)
```

```python
Ds = [2, 4, 8, 16, 32, 64]
targets = {D: Banana(dim=D, sigma=5.0, b=0.02, device=device) for D in Ds}
```

```python
def get_model(D, hidden_dims):
    latent_dim = D
    return VAE(data_dim=D, hidden_dims=hidden_dims, latent_dim=latent_dim, device=device)

def get_and_load_model(D, std_factor=1, hidden_dims=(128, 128, 128, 128)):
    model = get_model(D, hidden_dims)
    model.load_knowledge('../models/' + get_filename(model, targets[D]))
    model.set_std_factor(std_factor)
    return model
```

```python
def plot_mh_samples_for_all_Ds(N, L, std_factor, q=0, use_reeval=False, use_multiple=False):
    proj_dims = (0, 1)
    alpha = 0.3

    ncols = 3 + use_reeval + use_multiple
    nrows = len(Ds)
    fig, axss = plt.subplots(ncols=ncols, nrows=nrows, figsize=(4 * ncols, 4 * nrows))
    fig.suptitle(f'MH Correction of VAE Samples (Projections)\n std_factor={std_factor}; q={100*q:0.1f}%; L={L}; {N} samples', y=1.0)
    fig.tight_layout()
    
    for D, axs in tqdm(zip(Ds, axss)):
        target = targets[D]
        model = get_and_load_model(D, std_factor=1)
        proposer = lambda sample_shape : model.sample(sample_shape)
        proposal_log_prob_estimator = lambda x : model.iw_log_marginal_estimate(x, L=L, batch_L=32)

        target_samples = target.sample((N,))
        #stv_estimator = lambda samples : sliced_total_variation(target_samples, samples, n_kde_samples=300, n_projections=2*D)
        metric = WassersteinMetric()
        metric_estimator = lambda samples : metric(target_samples, samples)
        
        # Generate model samples
        _start_time = time()
        model_samples = model.sample((N,))
        model_time = time() - _start_time
        model_atv = metric_estimator(model_samples)

        # Remove unlikely samples
        cutoff_quantile =  target.log_prob(target_samples).quantile(q).item()
        _start_time = time()
        cut_acc_rate, cut_indicies = log_prob_cutoff_filter(target, model_samples, cutoff_quantile)
        cut_samples = model_samples[cut_indicies]
        cutoff_time = time() - _start_time
        cut_atv = metric_estimator(cut_samples)

        # Run Metropolis-Hastings algorithm
        _start_time = time()
        mh_acc_rate, mh_samples = metropolis_hastings_filter(target, cut_samples, proposal_log_prob_estimator)
        mh_time = time() - _start_time
        mh_atv = metric_estimator(mh_samples)

        # Run Metropolis-Hastings algorithm using multiple estimates
        if use_multiple:
            n_estimates = 10
            _start_time = time()
            mhm_acc_rate, mhm_samples = metropolis_hastings_filter(target, cut_samples, proposal_log_prob_estimator, n_estimates=n_estimates)
            mhm_time = time() - _start_time
            mhm_atv = metric_estimator(mhm_samples)

        # Plot
        ax1, ax2, ax3 = axs[:3]

        ax1.scatter(*pl(target_samples[:, proj_dims]), alpha=alpha, label='Target Samples')
        ax1.scatter(*pl(model_samples[:, proj_dims]), alpha=alpha, label='VAE Samples')
        ax1.set_title(f'D={D}; WM:{model_atv: 0.2f}; {model_time: 0.1f} sec')

        ax2.scatter(*pl(target_samples[:, proj_dims]), alpha=alpha, label='Target samples')
        ax2.scatter(*pl(cut_samples[:, proj_dims]), alpha=alpha, label='Cut VAE samples')
        ax2.set_title(f'D={D};  WM:{cut_atv: 0.2f}; Cut A/R={100 * cut_acc_rate: 0.1f}%; {cutoff_time: 0.1f} sec')  # cutoff time is currently 0.0 sec
        
        ax3.scatter(*pl(target_samples[:, proj_dims]), alpha=alpha, label='Target samples')
        ax3.scatter(*pl(mh_samples[:, proj_dims]), alpha=alpha, label='Corrected (no re-eval) VAE samples')
        ax3.set_title(f'D={D}; WM:{mh_atv: 0.2f}; A/R={100 * mh_acc_rate: 0.1f}%; {mh_time: 0.1f} sec')
        
        if use_multiple:
            ax4 = axs[3]
            ax4.scatter(*pl(target_samples[:, proj_dims]), alpha=alpha, label='Target Samples')
            ax4.scatter(*pl(mhm_samples[:, proj_dims]), alpha=alpha, label=f'Corrected ({n_estimates} estimates) VAE samples')
            ax4.set_title(f'D={D}; WM:{mhm_atv: 0.2f}; A/R={100 * mhm_acc_rate: 0.1f}%; {mhm_time: 0.1f} sec')

        for ax in axs:
            ax.set_xlim(ax2.get_xlim())
            ax.set_ylim(ax2.get_ylim())
            ax.set_xticks([])
            ax.set_yticks([])
            ax.legend()

plot_mh_samples_for_all_Ds(N=2000, L=64, std_factor=1, q=0.001, use_reeval=False, use_multiple=True)
```

```python
def sample_target_densities():
    use_gaussian = False
    
    D = 64
    std_factor = 1
    L = 512
    beta = 1
    N = 3000
    n_estimates = 10
    tv1d = TotalVariation1d(n_density_samples=1000)
    
    model = get_and_load_model(D, std_factor, hidden_dims=(128,128))
    target = targets[D]
    
    target_samples = target.sample((N,))
    if use_gaussian:
        gaussian_approx = approximate_with_gaussian(target_samples)
        model_samples = gaussian_approx.sample((N,))
        proposal_log_prob_estimator = lambda x :gaussian_approx.log_prob(x)
    else:
        model_samples = model.sample((N,))
        proposal_log_prob_estimator = lambda x : model.iw_log_marginal_estimate(x, L=L, beta=beta, batch_L=32)
    target_log_likelihood = target.log_prob(target_samples)
    proposal_log_likelihood = target.log_prob(model_samples)
    
    q_min = 0.001
    q_max = 0.999
    target_min_quantile = target_log_likelihood.quantile(q_min).item()
    target_max_quantile = target_log_likelihood.quantile(q_max).item()
    cut_acc_rate, cut_indicies = log_prob_cutoff_filter(target, model_samples, target_min_quantile)
    cut_proposal_samples = model_samples[cut_indicies]
    cut_proposal_log_likelihood = target.log_prob(cut_proposal_samples)
    max_acc_rate = 1 - tv1d(proposal_log_likelihood, target_log_likelihood)
    max_acc_rate_with_cut = 1 - tv1d(cut_proposal_log_likelihood, target_log_likelihood)

    mh_ar, mh_samples = metropolis_hastings_filter(target, cut_proposal_samples, proposal_log_prob_estimator, n_estimates=10)
    mh_log_likelihood = target.log_prob(mh_samples)

    ll_cutoff = -80
    
    fig, ax = plt.subplots(figsize=(16, 8))
    ax.set_xlim(left=ll_cutoff)

    # proposal_log_prob = proposal_log_prob_estimator(target_samples)
    ax.hist(
        list(map(lambda x: to_numpy(x[x > ll_cutoff]), (target_log_likelihood, proposal_log_likelihood, mh_log_likelihood))),
        bins=100,
        histtype='stepfilled',
        alpha=0.5,
        label=['Target samples', 'Proposal samples', 'MH samples'],
        density=True
    )
    ax.axvline(target_min_quantile, label=f"Target samples' {100*q_min}% log-density quantile")
    #ax.axvline(target_max_quantile, label=f"Target samples' {100*q_max}% log-density quantile", color='tab:red')
    #sns.histplot(to_numpy(proposal_log_likelihood), ax=ax, label='Proposal samples')
    
    #proposal_log_prob = proposal_log_prob_estimator(model_samples)
    #target_log_prob = target.log_prob(model_samples)
    #ax.scatter(to_numpy(target_log_prob.exp()), to_numpy(proposal_log_prob.exp()), alpha=0.5, label='Model samples')
    
    # ax.set_xscale('log')
    ax.set_xlabel('Log-likelihood under target distribution')
    #ax.set_ylabel('Proposal density')
    ax.set_title(f'Distribution of Log-likelihood Under Target Distribution\nD={D}; std_factor={std_factor}; L={L}; n_estimates={n_estimates}\n Theoretical maximum acceptance rate: {max_acc_rate * 100:0.0f}% ({max_acc_rate_with_cut * 100:0.0f}% with cut)')
    if use_gaussian:
        ax.set_title('GAUSSIAN Model\n' + ax.get_title())
    ax.legend()

sample_target_densities()
```

```python
def show_log_prob_outliers():
    D = 32
    std_factor = 2
    N = 1000
    proj_dims = (0, 1)
    alpha = 0.3
    
    model = get_and_load_model(D, std_factor, hidden_dims=(128, 128, 128, 128))
    target = targets[D]
    
    target_samples = target.sample((N,))
    model_samples = model.sample((N,))
    target_log_likelihood = target.log_prob(target_samples)
    model_log_likelihood = target.log_prob(model_samples)
    
    q_min = 0.001
    q_max = 0.999
    target_min_quantile = target_log_likelihood.quantile(q_min).item()
    target_max_quantile = target_log_likelihood.quantile(q_max).item()
    # cut_acc_rate, cut_indicies = log_prob_cutoff_filter(target, model_samples, target_min_quantile, target_max_quantile)
    low_prob_outliers = model_samples[model_log_likelihood < target_min_quantile]
    high_prob_outliers = model_samples[model_log_likelihood > target_max_quantile]
    n_low = low_prob_outliers.shape[0]
    n_high = high_prob_outliers.shape[0]
    
    fig, axs = plt.subplots(figsize=(16, 8), ncols=2)
    ax1, ax2 = axs

    ax1.scatter(*pl(target_samples[:, proj_dims]), alpha=alpha)
    ax1.set_title('Target Samples')
    
    ax2.scatter(*pl(low_prob_outliers[:, proj_dims]), alpha=alpha, label=f'Low-probability outliers ({100 * n_low / N :0.1f}%)', color='blue')
    ax2.scatter(*pl(high_prob_outliers[:, proj_dims]), alpha=alpha, label=f'High-probability outliers ({100 * n_high / N :0.1f}%)', color='red')
    ax2.set_title('Model Outliers')
    ax2.set_xlim(ax1.get_xlim())
    ax2.set_ylim(ax1.get_ylim())
    ax2.legend()

    fig.suptitle(f'Log-Probability Outliers (Projection)\nD={D}; std_factor={std_factor}')
    #plt.tight_layout()

show_log_prob_outliers()
```

```python
def choose_std_factor():
    D = 8
    L = 64
    beta = 1
    N = 2000
    #metric = SlicedDistributionMetric(TotalVariation1d(n_density_samples=100), n_projections=2 * D)
    metric = WassersteinMetric()
    
    model = get_and_load_model(D)
    target = targets[D]
    proposal_log_prob_estimator = lambda x : model.iw_log_marginal_estimate(x, L=L, beta=beta, batch_L=32)
    
    cutoff_quantile = 0.001
    target_samples = target.sample((N,))
    cutoff_log_prob = target.log_prob(target_samples).quantile(cutoff_quantile)
    
    std_factor_options = np.linspace(0.6, 1.4, 20)
    cutoff_acc_rates = []
    mh_acc_rates = []
    mh_atvs = []
    model_atvs = []
    for std_factor in tqdm(std_factor_options):
        model.set_std_factor(std_factor)
        proposal_samples = model.sample((N,))
        model_atvs.append(metric(target_samples, proposal_samples).item())
        cutoff_acc_rate, cut_indicies = log_prob_cutoff_filter(target, proposal_samples, cutoff_log_prob)
        cut_proposal_samples = proposal_samples[cut_indicies]
        mh_acc_rate, mh_samples = metropolis_hastings_filter(target, cut_proposal_samples, proposal_log_prob_estimator)
        mh_atvs.append(metric(target_samples, mh_samples).item())
        cutoff_acc_rates.append(cutoff_acc_rate)
        mh_acc_rates.append(mh_acc_rate)
                                      
    cutoff_acc_rates = np.array(cutoff_acc_rates)
    mh_acc_rates = np.array(mh_acc_rates)
    overall_acc_rates = cutoff_acc_rates * mh_acc_rates

    fig, axs = plt.subplots(figsize=(16, 16), nrows=2)
    ax = axs[0]
    ax.plot(std_factor_options, cutoff_acc_rates, label='Quantile cutoff acceptance rate')
    ax.plot(std_factor_options, mh_acc_rates, label='MH acceptance rate')
    ax.plot(std_factor_options, overall_acc_rates, label='Overall acceptance rate')
    ax.axhline(0, color='black', linestyle='--')
    ax.set_xlabel('std_factor')
    ax.legend()
    ax = axs[1]
    ax.plot(std_factor_options, model_atvs, label=f'{metric.name()} between target and model samples')
    ax.plot(std_factor_options, mh_atvs, label=f'{metric.name()} between target and MH samples')
    ax.set_xlabel('std_factor')
    ax.legend()
    plt.suptitle(f'Effect of std_factor on Sample Quality\nD={D}; L={L}; beta={beta}; N={N} cutoff_quantile={100 * cutoff_quantile}%', y=0.95)

choose_std_factor()
```

```python
# %%time

def atv_benchmark():
    L = 64
    beta = 1
    N = 4000
    q = 0.0001
    std_factor = 1

    scores = defaultdict(list)
    ars = defaultdict(lambda: np.zeros(len(Ds)))
    for i, D in enumerate(tqdm(Ds)):
        target = targets[D]
        model = get_and_load_model(D, std_factor)
        
        target_samples = target.sample((N,))
        target_log_likelihood = target.log_prob(target_samples)
        target_quantile = target_log_likelihood.quantile(q)

        metric = SlicedDistributionMetric(WassersteinMetric1d(), 3 * target.dim)
        metric_estimator = lambda samples: metric(target_samples, samples).item()

        gaussian_approx = approximate_with_gaussian(target_samples)
        gaussian_samples = gaussian_approx.sample((N,))
        scores['gaussian'].append(metric_estimator(gaussian_samples))
        ars['cut_gaussian'][i], cut_indicies = log_prob_cutoff_filter(target, gaussian_samples, target_quantile)
        cut_gaussian_samples = gaussian_samples[cut_indicies]
        scores['cut_gaussian'].append(metric_estimator(cut_gaussian_samples))
        gaussian_log_prob = lambda x : gaussian_approx.log_prob(x)
        ars['mh_gaussian'][i], mh_samples = metropolis_hastings_filter(target, cut_gaussian_samples, gaussian_log_prob)
        scores['mh_gaussian'].append(metric_estimator(mh_samples))

        vae_log_prob_estimator = lambda x : model.iw_log_marginal_estimate(x, L=L, beta=beta, batch_L=32)
        vae_samples = model.sample((N,))
        scores['vae'].append(metric_estimator(vae_samples))
        ars['cut_vae'][i], cut_indicies = log_prob_cutoff_filter(target, vae_samples, target_quantile)
        cut_vae_samples = vae_samples[cut_indicies]
        scores['cut_vae'].append(metric_estimator(cut_vae_samples))
        ars['mh_vae'][i], mh_samples = metropolis_hastings_filter(target, cut_vae_samples, vae_log_prob_estimator)
        scores['mh_vae'].append(metric_estimator(mh_samples))

        perfect_vae_samples = model.reconstruct(target.sample((N,)))
        scores['perfect_vae'].append(metric_estimator(perfect_vae_samples))
        scores['target'].append(metric_estimator(target.sample((N,))))

    #baseline = np.array(scores['target'])
    #for sample, score in scores.items():
    #    scores[sample] = np.array(score) / baseline
    
    fig, axs = plt.subplots(figsize=(10,12), nrows=2)
    fig.suptitle(f'Scaling with Dimensionality\nN={N}; std_factor={std_factor}; L={L}', y=0.95)
    
    ax = axs[0]
    plotstyle = {
        'perfect_vae': ('Perfect VAE samples', 'tab:purple', '-'),
        'target': ('Target samples', 'tab:red', '-'),
        'gaussian': ('Gaussian samples', 'tab:blue', ':'),
        'cut_gaussian': (f'Gaussian samples with {100*q:0.2f}% cutoff', 'tab:blue', '--'),
#        'mh_gaussian': ('MH with Gaussian proposal', 'tab:blue', '-'),
        'vae': ('VAE samples', 'tab:green', ':'),
        'cut_vae': (f'VAE Samples with {100*q:0.2f}% cutoff', 'tab:green', '--'),
        'mh_vae': ('MH with VAE proposal', 'tab:green', '-')
    }
    for sample_type, (label, color, linestyle) in plotstyle.items():
        ax.plot(Ds, scores[sample_type], label=label, color=color, linestyle=linestyle)
    ax.set_xlabel('D')
    ax.set_ylabel(f'{metric.name()} with target')
    ax.legend()
    
    ax = axs[1]
    ax.plot(Ds, ars['cut_gaussian'], label='Gaussian Cutoff', color='tab:blue', linestyle='--')
    ax.plot(Ds, ars['mh_gaussian'], label='Gaussian Proposal MH', color='tab:blue')
    ax.plot(Ds, ars['cut_vae'], label='VAE Cutoff', color='tab:green', linestyle='--')
    ax.plot(Ds, ars['mh_vae'], label='VAE Proposal MH', color='tab:green')
    ax.axhline(0, color='black', linestyle=':')
    ax.set_xlabel('D')
    ax.set_ylabel('Acceptance Rate')
    ax.legend()

atv_benchmark()
```

```python
def analyze_gaussian_proposal_atv():
    q = 0.0
    sigma = 6
    b = 0.03
    N = 4000
    alpha = 0.3
    Ds = np.arange(2, 66, 2)
    targets = {D: Banana(dim=D, sigma=sigma, b=b) for D in Ds}

    fig, axs = plt.subplots(figsize=(18, 12), ncols=3, nrows=2)
    fig.suptitle(f'Gaussian vs. Banana\nsigma={sigma}; b={b}; q={100*q}%; N={N}', y=1)

    gaussian_atvs = []
    cut_gaussian_ars = []
    cut_gaussian_atvs = []
    mh_gaussian_ars = []
    mh_gaussian_atvs = []
    self_atvs = []
    
    for D in tqdm(Ds):
        target = targets[D]
        target_samples = target.sample((N,))
        target_log_likelihood = target.log_prob(target_samples)
        target_quantile = target_log_likelihood.quantile(q)
        #tv_estimator = lambda samples : sliced_total_variation(target_samples, samples, n_kde_samples=300, n_projections=2*D, mode='max')
        metric = SlicedDistributionMetric(WassersteinMetric1d(), 3 * target.dim)
        metric_estimator = lambda samples: metric(target_samples, samples).item()
        
        gaussian_approx = approximate_with_gaussian(target_samples)
        gaussian_samples = gaussian_approx.sample((N,))
        gaussian_atvs.append(metric_estimator(gaussian_samples))
        acc_rate, cut_indicies = log_prob_cutoff_filter(target, gaussian_samples, target_quantile)
        cut_gaussian_ars.append(acc_rate)
        cut_gaussian_samples = gaussian_samples[cut_indicies]
        cut_gaussian_atvs.append(metric_estimator(cut_gaussian_samples))
        gaussian_log_prob = lambda x : gaussian_approx.log_prob(x)
        acc_rate, mh_samples = metropolis_hastings_filter(target, cut_gaussian_samples, gaussian_log_prob)
        mh_gaussian_ars.append(acc_rate)
        mh_gaussian_atvs.append(metric_estimator(mh_samples))
        self_atvs.append(metric_estimator(target.sample((N,))))

        if D == 10:
            proj_dims = (0, 1)
            proj_gaussian_samples = gaussian_samples[:, proj_dims]
            proj_target_samples = target_samples[:, proj_dims]
            ax = axs[0][0]
            ax.scatter(*pl(proj_gaussian_samples), alpha=0.3, label='Gaussian')
            ax.scatter(*pl(proj_target_samples), alpha=0.3, label='Target')
            ax.set_title(f'D={D}')
            ax = axs[0][1]
            ax.hist(
                [to_numpy(proj_gaussian_samples[:, 0]), to_numpy(proj_target_samples[:, 0])],
                bins=50,
                histtype='stepfilled',
                alpha=0.5,
                label=['Gaussian', 'Target'],
                density=True
            )
            x_atv = TotalVariation1d()(proj_target_samples[:, 0], proj_gaussian_samples[:, 0])
            ax.set_title(f'X-axis projection. TV={x_atv:0.3f}')
            ax = axs[0][2]
            ax.hist(
                [to_numpy(proj_gaussian_samples[:, 1]), to_numpy(proj_target_samples[:, 1])],
                bins=50,
                histtype='stepfilled',
                alpha=0.5,
                label=['Gaussian', 'Target'],
                density=True
            )
            y_atv = TotalVariation1d()(proj_target_samples[:, 1], proj_gaussian_samples[:, 1])
            ax.set_title(f'Y-axis projection. TV={y_atv:0.3f}')

    ax = axs[1][0]
    ax.plot(Ds, gaussian_atvs, label='Gaussian Samples', color='tab:blue', linestyle=':')
    ax.plot(Ds, cut_gaussian_atvs, label=f'Gaussian Samples with {100*q:0.2f}% cutoff', color='tab:blue', linestyle='--')
    ax.plot(Ds, mh_gaussian_atvs, label='MH with Gaussian Proposal', color='tab:blue')
    ax.plot(Ds, self_atvs, label='Target Samples', color='black')
    ax.set_xlabel('D')
    ax.set_ylabel(f'{metric.name()} with Target')

    ax = axs[1][1]
    ax.plot(Ds, cut_gaussian_ars, label='Gaussian Cutoff', color='tab:blue', linestyle='--')
    ax.plot(Ds, mh_gaussian_ars, label='Gaussian Proposal MH', color='tab:blue')
    ax.axhline(0, color='black', linestyle=':')
    ax.set_xlabel('D')
    ax.set_ylabel('Acceptance Rate')

    for axs_row in axs:
        for ax in axs_row:
            ax.legend()

analyze_gaussian_proposal_atv()
```

```python
def analyze_effect_of_beta():
    q = 0
    D = 32
    std_factor = 2
    L = 64
    beta_options = np.logspace(0, 1, 20)
    N = 2000
    metric = SlicedDistributionMetric(TotalVariation1d(), 3 * D)

    target = targets[D]
    target_samples = target.sample((N,))
    log_prob_cutoff = get_log_prob_quantile(target, q=q)
    stv_estimator = lambda samples : metric(target_samples, samples)
    model = get_and_load_model(D, std_factor)
    model_samples = model.sample((N,))
    cut_acc_rate, cut_indicies = log_prob_cutoff_filter(target, model_samples, log_prob_cutoff)
    cut_samples = model_samples[cut_indicies]
    cut_stv = stv_estimator(cut_samples[:N//10])

    mh_acc_rates = []
    stvs = []
    for beta in tqdm(beta_options):
        vae_log_prob_estimator = lambda x : model.iw_log_marginal_estimate(x, L=L, beta=beta, batch_L=64)
        mh_acc_rate, mh_samples = metropolis_hastings_filter(target, cut_samples, vae_log_prob_estimator)
        stv = stv_estimator(mh_samples)
        mh_acc_rates.append(mh_acc_rate)
        stvs.append(stv)

    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_xscale('log')
    ax.plot(beta_options, stvs, color='tab:red')
    ax.axhline(cut_stv, label='Cut samples STV', color='tab:red', linestyle='--')
    ax.legend()
    ax.set_xlabel('beta')
    ax.set_ylabel('Sliced total variation between target and MH samples', color='tab:red')
    ax2 = ax.twinx()
    ax2.plot(beta_options, mh_acc_rates, color='tab:blue')
    ax2.set_ylabel('MH acceptance rate', color='tab:blue')
    plt.title(f'Effect of beta on Sample Quality\nD={D}; L={L}; std_factor={std_factor}; N={N}; q={100 * q}%')

analyze_effect_of_beta()
```
